```python
# Q2. show SQL output
sql2 = '''
select Continent, count(*) As Number_Countries, sum(population) As Population from country
where population > 0
group by Continent
order by 1 ASC'''

sql_df2 = pd.read_sql_query(sql2, con = cnx)
sql_df2
```

- Description of SQL command: To select continents with their respective number of countries and totoal number of populations, with only continents having population > 0, and group these records by continent, then sort them in an ascending order.

```python
df2 = df_country[df_country.Population>0].groupby('Continent').agg(
    Number_Countries=pd.NamedAgg(column = 'Name', aggfunc = 'count'),
    Population=pd.NamedAgg(column = 'Population', aggfunc = 'sum')).sort_values('Continent', ascending=True).reset_index()
df2
```

- The output is the same for the two commands, except in pandas the continent names are sorted alphabetically, while in sql the output is sorted by ascii values of the strings.

```python
# Q4. show SQL output
sql4 = '''
select country.Name, Language, (Percentage * population) / 100
from countrylanguage
inner join country on countrylanguage.CountryCode = country.code
where IsOfficial = True
order by 3 DESC limit 10'''

sql_df4 = pd.read_sql_query(sql4, con = cnx)
sql_df4
```

- This SQL commmand intends to select the top 10 official languages that's used by most people and their country names, order by the number of people using it in a descending order.

```python
# Translate into Pandas Commands
df_countrylanguage=df_countrylanguage.reset_index()
df4 = df_countrylanguage.merge(df_country,how='inner',left_on='CountryCode',right_on='Code')
df4['(Percentage * population) / 100'] = (df4.Percentage * df4.Population) / 100
df4 = df4[df4.IsOfficial=='T'][['Name','Language','(Percentage * population) / 100']].sort_values('(Percentage * population) / 100', ascending=False).head(10).reset_index().drop(columns = 'index')
df4
```

- The output is the same

```python
# Q5. show SQL output
sql5 = '''
select Language, sum((Percentage * population) / 100)
from countrylanguage
inner join country ON countrylanguage.CountryCode = country.code
group by Language
order by 2 desc limit 5'''

sql_df5 = pd.read_sql_query(sql5, con = cnx)
sql_df5
```

- This SQL command intends to select the top 5 languages with most people using it, order by the number of people using it decreasingly

```python
df5 = df_countrylanguage.merge(df_country,how='inner',left_on='CountryCode',right_on='Code')
df5 = df5.groupby('Language').apply(lambda row: pd.Series({ "sum((Percentage * population) / 100)": ((row["Percentage"]*row["Population"])/100).sum() })).sort_values("sum((Percentage * population) / 100)", ascending = False).reset_index().head(5)
df5
```

- The output are the same
